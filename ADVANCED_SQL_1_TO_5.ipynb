Q1. What is a Common Table Expression (CTE), and how does it improve SQL query readability?

Ans.- A Common Table Expression (CTE) is a temporary result set defined using the WITH clause that exists only for the duration of a query.

 How it improves readability:

- Breaks complex queries into smaller logical parts
- Avoids repeating subqueries
- Makes code easier to debug and maintain

Q2. Why are some views updatable while others are read-only? Explain with an example.

Ans.- A view is updatable if:

- It is based on a single table
No aggregate functions (SUM, AVG, etc.)
No GROUP BY, DISTINCT, or JOIN

A view is read-only if:

- It uses joins, aggregates, or grouping

Example:

✅ Updatable view:

CREATE VIEW SimpleView AS
SELECT ProductID, ProductName, Price FROM Products;

❌ Read-only view:

CREATE VIEW ComplexView AS
SELECT Category, COUNT(*) FROM Products GROUP BY Category;

Q3. What advantages do stored procedures offer compared to writing raw SQL queries repeatedly?

Ans.
- Performance: Precompiled, faster execution

- Reusability: Write once, use many times

- Security: Restrict direct table access

- Maintainability: Centralized logic

- Reduced network traffic: Execute multiple queries in one call

Q4. What is the purpose of triggers in a database? Mention one use case where a trigger is essential.

Ans.- Triggers are automatic actions executed when events occur (INSERT, UPDATE, DELETE).

Use case:

- Audit logging (track deleted records)
- Enforcing business rules

Example: Automatically log deleted products.

Q5. Explain the need for data modelling and normalization when designing a database.

Ans.- Data Modeling:

- Defines structure of database (tables, relationships)

Normalization:

- Organizes data to reduce redundancy and improve integrity

Benefits:

- Eliminates duplicate data
- Ensures consistency
- Improves query performance
